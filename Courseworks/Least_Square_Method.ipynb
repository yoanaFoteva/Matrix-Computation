##Task 1

Find the normal solution of the least squares system:

$$
\begin{cases}
2x_1 + 3x_2 - 4x_3 = 1.2 \\
3x_1 + x_2 - 7x_3 = 4.7 \\
x_1 - 4x_2 - 3x_3 = 6.1 \\
4x_1 + 3x_2 - 5x_3 = 5.2 \\
2x_1 - 4x_2 + 3x_3 = 7.9
\end{cases}
$$

```python
import sympy as sp

A = sp.Matrix([
    [2, 3, -4],
    [3, 1, -7],
    [1, -4, -3],
    [4, 3, -5],
    [2, -4, 3]
])

b = sp.Matrix([
    sp.Rational(12,10),
    sp.Rational(47,10),
    sp.Rational(61,10),
    sp.Rational(52,10),
    sp.Rational(79,10)
])

x = (A.T*A).LUsolve(A.T*b)
x
```

```python
import sympy as sp

x1, x2, x3 = sp.symbols('x1 x2 x3')

r1 = 2*x1 + 3*x2 - 4*x3 - 1.2
r2 = 3*x1 + x2 - 7*x3 - 4.7
r3 = x1 - 4*x2 - 3*x3 - 6.1
r4 = 4*x1 + 3*x2 - 5*x3 - 5.2
r5 = 2*x1 - 4*x2 + 3*x3 - 7.9

f = r1**2 + r2**2 + r3**2 + r4**2 + r5**2

eq1 = sp.diff(f, x1)
eq2 = sp.diff(f, x2)
eq3 = sp.diff(f, x3)

solution = sp.solve([eq1, eq2, eq3], [x1, x2, x3])
solution
```

##Task 2

Given data points $(x_i, y_i)$, find the least squares estimates of the coefficients $a$ and $b$ for the model $y = a + \frac{b}{x}$.

| x   | 0.5 |  1  |  1.5 |   2  |  2.5 |
|-----|-----|-----|------|------|------|
| y   | 10.5| 1.6 | 0.55 | 0.26 | 0.13 |

```python
import numpy as np

x = np.array([0.5, 1, 1.5, 2, 2.5], dtype=float)
y = np.array([10.5, 1.6, 0.55, 0.26, 0.13], dtype=float)

A = np.column_stack([np.ones_like(x), 1/x])

theta, residuals, rank, s = np.linalg.lstsq(A, y, rcond=None)
a, b = theta

print("a =", a)
print("b =", b)
```
